```python
my_data=[43014687,22427,42475588,516672]
my_labels=['Confirmed','Active','Cured/Discharged','Death']
my_explode=(0,0,0,1.5)
my_colors=['green','blue','pink','black']
#fig1 = plt.figure(figsize(13,8))
#ax1=fig.add_subplot(111)
plt.pie(my_data,labels=my_labels,autopct='%1.1f%%',startangle=90,shadow=True,colors=my_colors,explode=my_explode)
plt.axis('equal')
plt.title('Covid-19 Cases in India')
plt.show()
```
